```python
import gymnasium as gym
import time

env = gym.make("CartPole-v1", render_mode="human")

print("Action Space:", env.action_space)
print("  - 0: Push cart to the LEFT")
print("  - 1: Push cart to the RIGHT")
print("observation space: [cart position, cart velocity, pole angle, pole angular velocity]")
print("Observation Space:", env.observation_space)
print()

obs, info = env.reset()
episode = 1
episode_reward = 0



def get_action(obs):
    cart_pos, cart_vel, pole_angle, pole_ang_vel = obs

    # Extra if theres more steps allowed per episode
    # Keep cart close to center
    if abs(cart_pos) > 1.5:
        return 0 if cart_pos > 0 else 1

    # Predict future angle using angular velocity
    future_angle = pole_angle + 0.2 * pole_ang_vel

    # Main condition move in direction of predicted angle
    if abs(future_angle) < 0.02:  # Nearly vertical
        return 0  # Default to left when stable
    else:
        return 1 if future_angle > 0 else 0

for episode in range(1, 20):
  for step in range(500):
    action = get_action(obs)  # New action
    # env.step(action) returns 5 values:
    #   obs        - Observation: array of 4 floats describing the current state:
    #                  [0] Cart Position      (-4.8 to 4.8, episode ends at 2.4)
    #                  [1] Cart Velocity       (how fast the cart moves left/right)
    #                  [2] Pole Angle          (radians, episode ends at 12)
    #                  [3] Pole Angular Vel.   (how fast the pole is rotating)
    #   reward     - Always 1.0 for each step the pole stays up.
    #                Total episode reward = number of steps survived.
    #   terminated - True if the episode ended due to failure:
    #                  pole angle > 12(0.2094 radians) or cart position > 2.4
    #   truncated  - True if the episode ended due to time limit (500 steps).
    #                This means the agent survived the full episode (best outcome).
    #   info       - Dict with extra diagnostic info (usually empty in CartPole).
    obs, reward, terminated, truncated, info = env.step(action)

    episode_reward += reward
    print(f"Step {step} | Action: {action} | Reward: {reward} | Obs: {obs}")
    if terminated:
        print()
        print(f" --> Episode {episode} TERMINATED (pole fell or went out of bounds) | Total Episode Reward: {episode_reward}")
        print()
        episode_reward = 0
        time.sleep(3)
        obs, info = env.reset()
        break
    elif truncated:
        print()
        print(f"  --> Episode {episode} TRUNCATED (max steps reached) | Total Episode Reward: {episode_reward}")
        print()
        episode_reward = 0
        obs, info = env.reset()
        break

env.close()
```

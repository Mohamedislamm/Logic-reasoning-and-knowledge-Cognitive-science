```python
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1

```python
a = [1, 5, 9, 7, 5, 3]
```

```python
mean = np.mean(a)
median = np.median(a)
variance = np.var(a)
std_dev = np.std(a)
print(f"List: {a}")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# From scratch
# Mean
mean = sum(a)/len(a)
print(f"Mean = sum/count = {sum(a)}/{len(a)} = {mean:.1f}")

# Median
s = sorted(a)
n = len(a)
median = s[n//2] if n%2 else (s[n//2-1]+s[n//2])/2
print(f"Median = middle of {s} = {median}")

# Population Variance
var = sum((x-mean)**2 for x in a)/len(a)
print(f"Variance = avg squared diff = {var:.1f}")

# Standard Deviation
std = var**0.5
print(f"Std Dev = sqrt(var) = {std:.1f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import poisson
```

##Problem 1

```python
# Generate 1000 random heights (mean=175, std=6)
heights = norm.rvs(loc=175, scale=6, size=1000)
```

```python
# Plot the histogram with the normal curve
plt.figure(figsize=(8, 5))
plt.hist(heights, bins=25, density=True, alpha=0.6, color='blue', label='Data')
```

```python
# Plot the theoretical normal distribution
x = np.linspace(150, 200, 100)
plt.plot(x, norm.pdf(x, 175, 6), 'r-', lw=2, label='Normal Dist')
plt.title("Height Distribution of Men (cm)")
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# Print statistics
print(f"Mean height: {np.mean(heights):.1f} cm")
print(f"Standard deviation: {np.std(heights):.1f} cm")
```

##Problem 2

```python
# Generate 10,000 years of dust storm data (=3.4)
storms = poisson.rvs(mu=3.4, size=10000)
```

```python
# Count frequency of each storm count and sort frequencies
unique, counts = np.unique(storms, return_counts=True)
sorted_data = sorted(zip(unique, counts))

# Unpack sorted data
storms, years = zip(*sorted_data)
```

```python
# Plot
plt.figure(figsize=(8, 5))
plt.bar(storms, years, color='orange', alpha=0.7)
plt.title('Dust Storms in Multan (10,000 years)')
plt.xlabel('Storms per year')
plt.ylabel('Number of years')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
